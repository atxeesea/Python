Задание 1

                Создайте программу,хранящую информацию о великих баскетболистах. Нужно хранить ФИО баскетболиста и его рост.Требуется реализовать возможность добавления, удаления, поиска, замены данных. Используйте словарь для хранения информации.

```python
nba_legends = {
    'Фио':['Стефен Карри'],
    'Рост':['1,88 м']}

def info(d):
    return print(d)

def delete_choose(key):
    del nba_legends[key]
    print('Все удалилось')

def add_info(my_dict,key,value):
    my_dict[key] = value
    print("информация была добавлеан")

def search_info(my_dict, key):
    return my_dict.get(key, "Ключ не найден.")

def replace_info(my_dict,key,value):
    if key in my_dict:
        my_dict[key] = value
        print('Успешно')

    else:
        print('ключ не найден')

def main():
    while True:
        print("\nВыберите действие")
        print("1.Показать информацию")
        print("2.Добавить информацию")
        print("3.Заменить ифнормацию")
        print("4.Найти информацию")
        print('5. Удаление')
        print('6.Выйти из программы')
    
        choose = input("Выберите номер действия")
        print(choose)
    
        if choose == '1':
            print(nba_legends)

        elif choose == '2':
            key = input('Введите ключ:')
            print(key)

            value = input("Введите значение:")
            print(value)

            add_info(nba_legends,key,value)

        elif choose == '3':
            key = input("Введите ключ для замены")
            print(key)

            value = input("Введите новое значение")
            print(value)

            replace_info(nba_legends,key,value)

        elif choose == '4':
            key = input("Введите ключ для поиска: ")
            print(search_info(nba_legends, key))

        elif choose =='5':
            key = input('Выберите какой ключ удалить')
            delete_choose(key)

        elif choose == '6':
            print("Выход из программы.")
            break
        
        else:
            print("Вы ввели не то ")

if __name__ == "__main__":
    main()
```
