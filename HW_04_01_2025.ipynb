Задание 1 <br>

Создайте класс Circle (окружность). Для данного
класса реализуйте ряд перегруженных операторов:
- Проверка на равенство радиусов двух окружностей(операция = =);
- Сравнения длин двух окружностей (операции >, <,<=,>=);
- Пропорциональное изменение размеров окружности,путем изменения ее радиуса (операции + - += -=).

```python
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def __eq__(self, other):
        if isinstance(other, Circle):
            return self.radius == other.radius
        return NotImplemented

    def __lt__(self, other):
        if isinstance(other, Circle):
            return self.radius < other.radius
        return NotImplemented

    def __gt__(self, other):
        if isinstance(other, Circle):
            return self.radius > other.radius
        return NotImplemented

    def __le__(self, other):
        if isinstance(other, Circle):
            return self.radius <= other.radius
        return NotImplemented

    def __ge__(self, other):
        if isinstance(other, Circle):
            return self.radius >= other.radius
        return NotImplemented

    def __add__(self, other):
        if isinstance(other, (int, float)):
            return Circle(self.radius + other)
        return NotImplemented

    def __sub__(self, other):
        if isinstance(other, (int, float)):
            return Circle(self.radius - other)
        return NotImplemented

    def __iadd__(self, other):
        if isinstance(other, (int, float)):
            self.radius += other
            return self
        return NotImplemented

    def __isub__(self, other):
        if isinstance(other, (int, float)):
            self.radius -= other
            return self
        return NotImplemented



circle1 = Circle(5)
circle2 = Circle(3)

circle1 += 2  
circle2 -= 1  

print(f'Радиус circle1: {circle1.radius}')  
print(f'Радиус circle2: {circle2.radius}')  

print(circle1 > circle2)
```

Создайте класс Complex (комплексное число). Более подробно ознакомиться с комплексными числами можно по ссылке. Создайте перегруженные операторы для реализации арифметических операций для по работе с комплексными числами (операции +, -, *, /).

```python
class Complex:
    def __init__(self,complex_num):
        self.complex_num = complex_num

# imag - Мнимая часть комплексного числа a + bi
# real - действительная часть числа
class Complex:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def __add__(self, other):
        return Complex(self.real + other.real, self.imag + other.imag)

    def __sub__(self, other):
        return Complex(self.real - other.real, self.imag - other.imag)

    def __mul__(self, other):
        return Complex(self.real * other.real - self.imag * other.imag,
                       self.real * other.imag + self.imag * other.real)

    def __truediv__(self, other):
        denominator = other.real ** 2 + other.imag ** 2
        return Complex((self.real * other.real + self.imag * other.imag) / denominator,
                       (self.imag * other.real - self.real * other.imag) / denominator)

    def __str__(self):
        return f"{self.real} + {self.imag}i"

a = Complex(2, 3)
b = Complex(4, 5)

print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

Вам небходимо создать класс Airplane (самолет). С помощью перезагрузки операторов реализовать:
- Проверка на равенство типов самолетов (операция = =);
- Увеличение и уменьшение пассажиров в салоне самолета (операции + - += -=);
- Сравнение двух самолетов по максимально возможному количеству пассажиров на борту (операции > < <= >=).

```python
class Airplane:
    def __init__(self, type_airplane, passengers, max_passengers):
        self.type_airplane = type_airplane
        self.passengers = passengers
        self.max_passengers = max_passengers
    
    def __eq__(self, other):
        if isinstance(other, Airplane):
            return self.type_airplane == other.type_airplane
        return NotImplemented

    def __lt__(self, other):
        if isinstance(other, Airplane):
            return self.max_passengers < other.max_passengers
        return NotImplemented

    def __gt__(self, other):
        if isinstance(other, Airplane):
            return self.max_passengers > other.max_passengers
        return NotImplemented

    def __le__(self, other):
        if isinstance(other, Airplane):
            return self.max_passengers <= other.max_passengers
        return NotImplemented

    def __ge__(self, other):
        if isinstance(other, Airplane):
            return self.max_passengers >= other.max_passengers
        return NotImplemented

    def __add__(self, other):
        if isinstance(other, (int, float)):
            return Airplane(self.passengers + other)
        return NotImplemented

    def __sub__(self, other):
        if isinstance(other, (int, float)):
            return Airplane(self.passengers - other)
        return NotImplemented

    def __iadd__(self, other):
        if isinstance(other, (int, float)):
            self.passengers += other
            return self
        return NotImplemented

    def __isub__(self, other):
        if isinstance(other, (int, float)):
            self.passengers -= other
            return self
        return NotImplemented

airplane1 = Airplane("Boeing 737", 150, 200)
airplane2 = Airplane("Airbus A320", 180, 220)

print(airplane1 == airplane2)  
print(airplane1 < airplane2)   
airplane1 += 10
print(airplane1.passengers)
```

Задание 4 <br>

Создать класс Flat (квартира). Реализовать перегруженные операторы:
- Проверка на равенство площадей квартир (операция = =);
- Проверка на неравенство площадей квартир (операция !=);
- Сравнение двух квартир по цене (операции > < <= >=).

```python
class Flat:
    def __init__(self, square, price):
        self.square = square
        self.price = price
    
    def __eq__(self, other):
        if isinstance(other, Flat):
            return self.square == other.square
        return NotImplemented

    def __ne__(self, other):
        return not self == other

    def __lt__(self, other):
        if isinstance(other, Flat):
            return self.price < other.price
        return NotImplemented

    def __le__(self, other):
        if isinstance(other, Flat):
            return self.price <= other.price
        return NotImplemented

    def __gt__(self, other):
        if isinstance(other, Flat):
            return self.price > other.price
        return NotImplemented

    def __ge__(self, other):
        if isinstance(other, Flat):
            return self.price >= other.price
        return NotImplemented

    def __add__(self, amount):
        if isinstance(amount, (int, float)):
            return Flat(self.square, self.price + amount)
        return NotImplemented

    def __sub__(self, amount):
        if isinstance(amount, (int, float)):
            new_price = self.price - amount
            if new_price < 0:
                return Flat(self.square, new_price)
        return NotImplemented

    def __iadd__(self, amount):
        if isinstance(amount, (int, float)):
            self.price += amount
            return self
        return NotImplemented

    def __isub__(self, amount):
        if isinstance(amount, (int, float)):
            new_price = self.price - amount
            if new_price < 0:
                self.price = new_price
            return self
        return NotImplemented

    def __repr__(self):
        return f"Flat(square={self.square}, price={self.price})"

flat1 = Flat(50, 100000)
flat2 = Flat(75, 150000)

print(flat1 == flat2)
print(flat1 != flat2)
print(flat1 < flat2)


flat1 += 5000
print(flat1)

flat1 -= 10000
print(flat1)
```
