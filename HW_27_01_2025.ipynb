Задание 1

Пользователь вводит с клавиатуры набор чисел. По лученные числа необходимо сохранить в список (тип списка нужно выбрать в зависимости от поставленной ниже задачи).После чего нужно показать меню в котором предложить пользователю набор пунктов:
 1. Добавитьновоечисловсписок(еслитакоечислосу
ществует в списке, нужно вывести сообщение поль
зователю об этом, без добавления числа).
 2. Удалитьвсевхождениячислаизсписка(пользователь 
вводит с клавиатуры число для удаления)
 3. Показать содержимое списка (в зависимости от вы
бора пользователя список нужно показать с начала 
или с конца)
 4. Проверить есть ли значение в списке
 5. Заменить значение в списке (пользователь опреде
ляет заменить ли только первое вхождение или все 
вхождения)
 Взависимостиотвыборапользователявыполняется 
действие, после чего меню отображается снова.

```python
number_list = []
while True:
    a = input("Введите набор чисел (или 'exit' для выхода): ") 
    if a.lower() == 'exit':
        print("Программа завершена.")
        break

    print(a)
    numbers = a.split()
    for num in numbers:
        try:
            number = int(num)
            number_list.append(number)
            print(f'Ваше число {number} добавлено')
        except ValueError:
            print(f'"{num}" не число')

    print("Ваш список", number_list)

    while True:
        try:
            number_to_remove = int(input("Введите число для удаления из списка: "))
            number_list = [num for num in number_list if num != number_to_remove]
            print("Ваш новый список чисел", number_list)
            break
        except ValueError:
            print("Вы ввели не то")

    while True:
        choose = input("Вывод списка с начала - 1, а с конца - 2: ")
        if choose == '1':
            print("Список с начала:", number_list)
            break
        elif choose == '2':
            print("Список с конца:", number_list[::-1])
            break
        else:
            print("Вы ввели не то")

    while True:
        check = input("Введите число для проверки наличия: ")
        if check.lower() == 'exit':
            print("Программа завершена.")
            break
        try:
            check_number = int(check)
            if check_number in number_list:
                print(f'Число {check_number} имеется в списке.')
            else:
                print(f'Число {check_number} отсутствует в списке.')
        except ValueError:
            print("Вы ввели не то")
        value_to_replace = input("Введите число, которое хотите заменить или exit для выхода ")
        if value_to_replace.lower() == 'exit':
            print("Программа завершена.")
            break
        
        try:
            value_to_replace = int(value_to_replace)
            new_value = int(input("Введите новое значение "))
            replace_all = input("Заменить все вхождения? 1-да 2-нет: ")

            if replace_all == '1':
                number_list = [new_value if x == value_to_replace else x for x in number_list]
                print(f"Все вхождения {value_to_replace} заменены на {new_value}.")
            elif replace_all == '2':
                replaced = False
                for idx, x in enumerate(number_list):
                    if x == value_to_replace:
                        number_list[idx] = new_value
                        print(f"Первое вхождение {value_to_replace} заменено на {new_value}.")
                        replaced = True
                        break
                if not replaced:
                    print(f"Число {value_to_replace} не найдено в списке.")
            else:
                print("Вы ввели не то")
        except ValueError:
            print("Вы ввели не то")
```

Задание 2

Реализуйте класс стека для работысостроками(стек строк). Стек должен иметь фиксированный размер. Реализуйте набор операций для работы со стеком: 
- помещение строки в стек;
- выталкивание строки из стека;
- подсчет количества строк в стеке;
- проверку пустой ли стек;
- проверку полный ли стек;
- очистку стека;
- получение значения без выталкивания верхней строки 
из стека.

При старте приложения нужно отобразить меню с 
помощью, которого пользователь может выбрать необ
ходимую операцию.

```python
class Stack:
    def __init__(self, size):
        self.size = size
        self.stack = []
    
    def push(self, item):
        if len(self.stack) < self.size:
            self.stack.append(item)
            print(f"Строка '{item}' добавлена в стек.")
        else:
            print("стек полон, девозможно добавить строку.")
    
    def pop(self):
        if self.stack:
            return self.stack.pop()
        else:
            print("стек пуст, девозможно извлечь строку.")
            return None
    
    def count(self):
        return len(self.stack)
    
    def is_empty(self):
        return len(self.stack) == 0
    
    def is_full(self):
        return len(self.stack) == self.size
    
    def clear(self):
        self.stack.clear()
        print("стек очищен.")
    
    def up_element(self):
        if self.stack:
            return self.stack[-1]
        else:
            print("Стек пуст.")
            return None

def main():
    stack_size = int(input("Введите размер стека: "))
    stack = Stack(stack_size)
    
    while True:
        print("\nМеню:")
        print("1. помещение строки в стек")
        print("2. выталкивание строки из стека")
        print("3. подсчет количества строк в стеке")
        print("4. проверка пустой ли стек")
        print("5. проверка полный ли стек")
        print("6. очистка стека")
        print("7. получение значения без выталкивания верхней строки из стека")
        print("8. Выйти")
        
        choice = input("Выберите пункт меню: ")

        match choice: # можно было сделать через if-elif-else, но я решил поворить case, тк недавно использовал его решая задачи на c# и подумал, что эта конструкция будет приятнее на глаз
            case "1":
                item = input("Введите строку для добавления: ")
                stack.push(item)
        
            case "2":
                item = stack.pop()
                if item:
                    print(f"Извлечённая строка: {item}")
        
            case "3":
                print(f"Количество строк в строке: {stack.count()}")
        
            case "4":
                if stack.is_empty():
                    print("стек пуст")
                else:
                    print("стек не пуст")      
        
            case "5":
                if stack.is_full():
                    print("стек полон")
                else:
                    print("стек не полон")
        
            case "6":
                stack.clear()
        
            case "7":
                item = stack.up_element()
                if item:
                    print(f"Верхняя строка: {item}")
        
            case "8":
                print("Выход из программы.")
                break
        
            case _:
                 print("неверное значение")

if __name__ == "__main__":
    main()
```
