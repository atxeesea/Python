Задание 1

        Напишите информационную систему «Сотрудники». Программа должна обеспечивать ввод данных, редактирование данных сотрудника, удаление сотрудника, поиск сотрудника по фамилии, вывод информации обо всех сотрудниках, указанного возраста, или фамилия которых начинается на указанную букву. Организуйте возможность сохранения найденной информации в файл. Также весь список сотрудников сохраняется в файл (при выходе из программы — автоматически, в процессе исполнения программы — по команде пользователя). При старте программы происходит загрузка списка сотрудников из указанного пользователем файла.

```python
user_file = input()
try:
    with open(user_file, encoding='utf-8') as file:
        a = file.readlines()
except FileNotFoundError:
    print('Невозможно открыть файл')

while True:
    try:
        with open(user_file, encoding='utf-8') as file:
            s = file.read()
    except FileNotFoundError:
        print('Невозможно открыть файл')
    user_command = input()
    if user_command == '1':
        user_input = input('Ввод данных')
        try:
            with open(user_file, 'a', encoding='utf-8' ) as file:
                s = file.writelines(user_input)
        except FileNotFoundError:
            print('Невозможно открыть файл')
    
    elif user_command == '2':
        user_search = input('Поиск')
        user_input1 = input('Замена')
        s = s.replace(user_search, user_input1)
            
        try:
            with open(user_file, 'w' ,encoding='utf-8' ) as file:
                s = file.write(s)
        except FileNotFoundError:
            print('Невозможно открыть файл')

    elif user_command == '3':
        try:
            with open(user_file, encoding='utf-8') as file:
                s = file.readlines()
        except FileNotFoundError:
                print('Невозможно открыть файл')
        user_search = input('Поиск')
        w = []
        for i in range(len(s)):
            if user_search in s[i]:
                w.append(i)
        w = w[::-1]
        for i in w:
            s.pop(i)
        try:
            with open(user_file, 'w' ,encoding='utf-8' ) as file:
                s = file.writelines(s)
        except FileNotFoundError:
            print('Невозможно открыть файл')

    elif user_command == '4':

        user_search = input('Поиск')
        if user_search in s:
            print('Такой сотрудник есть')
        else:
            print('Нету')

    elif user_command == '5':
        print(s)

    elif user_command == '6':
        user_age = input()
        for i in a:
            if user_age in i:
                print(i)

    elif user_command == '7':
        try:
            with open(user_file, encoding='utf-8') as file:
                a = file.readlines()
        except FileNotFoundError:
            print('Невозможно открыть файл')
            user_letter = input()
        for i in a:
            if user_letter in i[0]:
                print(i)

    elif user_command == '0':
        break
```
